## General KNN

**Import necessary modules**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statistics import mode
from scipy.stats import rankdata
import math
%matplotlib inline
```

**Load Iris Dataset**

```python
df=sns.load_dataset('iris') 
```

```python
#SPLIT FEATURES AND LABELS

X=df.drop('species',axis=1)
y=df.species
```

```python
#SPLIT TRAIN AND TEST DATA

from  sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1004)
```

```python
#CONVERTING DATAFRAME INTO NUMPY_ARRAY FOR FASTER COMPUTATIONS

xtest=np.array(X_test)
xtrain=np.array(X_train)
xtrain.shape
```

**DEFINING  A  PREDICTION  FUNCTION  WITH  K_VALUE  AS  PARAMETER**

```python
def prediction(k):
    y=[(xtrain-xtest[i])**2 for i in range(len(xtest))] 
    c=np.zeros((len(xtest),len(xtrain)))
    for i in range(len(xtest)):
        for j in range(len(xtrain)):
            c[i][j]=np.array([np.sqrt(sum(y[i][j]))])  #CALCULATING THE DISTANCES AND STORE IT AS MATRICS
    
    predict=[]
    for i in range(len(xtest)):
        ind=np.argsort(c[i])                      #SORTING VALUES BASED ON INDICES
        k_labels=y_train.iloc[ind[:k]]            #SLICING VALUES BASED ON K_VALUE
        predict.append(mode(k_labels))            #SELECTING AND RETURNING MAXIMUM OCCURED LABEL
    return(predict)
    
```

**INITILIZING K_VALUE**

```python
k=1                                   
predict=prediction(k)                 #CALLING PREDICTION FUNCTION WITH K_VALUE AS ARGUMENTS
print(predict)
```

```python
from sklearn.metrics import accuracy_score
accurasy=accuracy_score(y_test,predict)        #CALCULAING ACCURACY
accurasy
```

**ERROR RATE**

```python
error_rate = 1-accurasy
print(f' IRIS Test Error_rate for k = {k} is {error_rate}')
```

## CONFORMAL PREDICTION  FOR IRIS

```python
#STORING UNIQUE LABELS TO SPECIES VARIEBLE
species=y_train.unique() 
species
```

**CALCULATING MINIMUM DISTANCE BETWEEN SAME_SPECIES WITHOUT INCLUDING TEST_DATA**

```python
# CALCULATING MINIMUM DISTANCE BETWEEN SAME SPECIES WITHOUT INCLUDING **TEST_DATA**
minimum=[]
for i in range(len(X_train)):
    x=[]
    for j in range(len(X_train)):
        if i!=j and y_train.iloc[i]==y_train.iloc[j]:
            ED=np.sqrt(sum((X_train.iloc[i]-X_train.iloc[j])**2))
            x.append(ED)
    minimum.append(min(x))
minimum1=minimum.copy()
minimum
```

**CALCULATING MINIMUM DISTANCE BETWEEN DIFFERENCE_SPECIES WITHOUT INCLUDING TEST_DATA**

```python
minimum2=[]
for i in range(len(X_train)):
    x1=[]
    for j in range(len(X_train)):
        if i!=j and y_train.iloc[i]!=y_train.iloc[j]:
            ED1=np.sqrt(sum((X_train.iloc[i]-X_train.iloc[j])**2))
            x1.append(ED1)
    minimum2.append(min(x1))
minimum3=minimum2.copy()
minimum2
```

**SAME_SPECIES AND DIFF_SPECIES FUNCTION WITH PARAMETERS - SPECIES_NAME, X_TEST DATA_POINT AND LIST OF PREVIOUSLY CALCULATED MINIMUM DISTANCE OF SAME_SPECIES AND DIFF_SPECIES**

```python
def same_species(minimum,xtest,species):
    z=np.where(y_train==species)                     # GETTING Y_TRAIN INDICES BASED ON SPECIES_NAME
    temp=X_train.iloc[np.where(y_train==species)]    #SPLITING THE DATA BASED ON SPECIES NAME
    y=(temp-xtest)**2                                
    ss=[]                    
    for i in range(len(y)):                          #CALCULATING DISTANCE BETWEEN X_TEST AND SPILITED DATA AND NOTE ITS MINIMUM.
        ss.append(np.sqrt(sum(y.iloc[i])))
        
    for i in range(len(y)):
        minimum[z[0][i]]=min(minimum[z[0][i]],ss[i])  #COMPARE AND OVERWRITE MIN BETWEEN CALCULATED DISTANCES 
    minimum.append(min(ss))                           #APPEND MINIMUM
    return(np.array(minimum))
```

```python
def diff_species(minimum,xtest,species):
    z=np.where(y_train!=species)                      # GETTING EVERY Y_TRAIN INDICES EXCEPT SPECIES_NAME VALUES
    temp=X_train.iloc[np.where(y_train!=species)]
    y=(temp-xtest)**2
    ss=[]
    for i in range(len(y)):
        ss.append(np.sqrt(sum(y.iloc[i])))
        
    for i in range(len(y)):
        minimum[z[0][i]]=min(minimum[z[0][i]],ss[i])
    minimum.append(min(ss))
    return(np.array(minimum))
```

**CALCULATE P_VALUE**

```python
p_value=np.zeros((len(X_test),len(species)))
for i in range(len(X_test)):
    for j in range(len(species)):
        minimum=minimum1.copy()
        c2=same_species(minimum,X_test.iloc[i],species[j])   #CALLING SAME_SPECIES FUNCTION
        minimum2=minimum3.copy()
        c1=diff_species(minimum2,X_test.iloc[i],species[j])  #CALLING DIFFERENT_SPECIES FUNCTION
        c_score=c1/c2                                        #CALULATING CONFORMITY SCORE (DIFF/SAME)
        p=rankdata(c_score,method='max')                     #CALCULATING RANK 
        p_value[i][j]=p[-1]/(len(X_train)+1)                 #CALCULATING P_VALUE OF X_TEST DATA (RANK/N+1)
```

**P_VALUE DATA_FRAME FOR SPECIES**

```python
df3=pd.DataFrame(p_value,columns=species)                     
df3
```

**AVERAGE FALSE P_VALUE CALULATION**

```python
for i in range(len(species)):
    v=np.where(y_test==species[i])
    df3[species[i]].iloc[v]=np.NaN
```

```python
avg_false_p_value=np.array(df3)
print(f'Average false P Value = {np.nanmean(avg_false_p_value)}')
```

# KNN FOR IONOSPHERE DATASET (GENERAL)

**READ IONOSPHERE DATASET**

```python
ion_df= pd.read_csv('ionosphere.txt')
```

```python
#SPLIT FEATURES AND LABELS
X1=ion_df.drop(ion_df.columns[-1],axis=1)
y1=ion_df[ion_df.columns[-1]]
```

```python
#SPLIT INTO TRAINING AND TESTING DATASET
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.3, random_state=1004)
```

```python
#CONVERTING DATAFRAME INTO NUMPY ARRAY
xtest1=np.array(X_test1)
xtrain1=np.array(X_train1)
xtrain1.shape
```

**DEFINING  A  PREDICTION  FUNCTION  WITH  K_VALUE  AS  PARAMETER**

```python
def prediction(k):
    y=[(xtrain1-xtest1[i])**2 for i in range(len(xtest1))] 
    c=np.zeros((len(xtest1),len(xtrain1)))
    for i in range(len(xtest1)):
        for j in range(len(xtrain1)):
            c[i][j]=np.array([np.sqrt(sum(y[i][j]))])
    
    predict=[]
    for i in range(len(xtest1)):
        ind=np.argsort(c[i])
        k_labels=y_train1.iloc[ind[:k]]
        predict.append(mode(k_labels))
    return(predict)
```

```python
k=1                       #INITIALIZE K_VALUE
predict1=prediction(k)    # CALL PREDICTION FUNCTION
print(predict1)
```

```python
accurasy1=accuracy_score(y_test1,predict1)        #CALCULAING ACCURACY
```

**ERROR RATE CALCULATION**

```python
error_rate1 = 1-accurasy1
print(f'IONOSPHERE Error_rate for k = {k} is {error_rate1}')
```

## CONFORMAL PREDICTION FOR IONOSPHERE DATASET

```python
#STORING UNIQUE LABELS
label=y_train1.unique()
label
```

**CALCULATING MINIMUM DISTANCE BETWEEN SAME_SPECIES WITHOUT INCLUDING TEST_DATA**

```python
mini=[]
for i in range(len(X_train1)):
    x1=[]
    for j in range(len(X_train1)):
        if i!=j and y_train1.iloc[i]==y_train1.iloc[j]:
            ED1=np.sqrt(sum((X_train1.iloc[i]-X_train1.iloc[j])**2))
            x1.append(ED1)
    mini.append(min(x1))
mini1=mini.copy()
mini1
```

**CALCULATING MINIMUM DISTANCE BETWEEN DIFFERENCE_SPECIES WITHOUT INCLUDING TEST_DATA**

```python
mini2=[]
for i in range(len(X_train1)):
    x2=[]
    for j in range(len(X_train1)):
        if i!=j and y_train1.iloc[i]!=y_train1.iloc[j]:
            ED2=np.sqrt(sum((X_train1.iloc[i]-X_train1.iloc[j])**2))
            x2.append(ED2)
    mini2.append(min(x2))
mini3=mini2.copy()
mini3
```

**SAME_LABEL AND DIFF_LABEL FUNCTION WITH PARAMETERS - LABEL_NAME, X_TEST DATA_POINT AND LIST OF PREVIOUSLY CALCULATED MINIMUM DISTANCE OF SAME_LABEL AND DIFF_LABEL**

```python
def same_label1(minimum,xtest,label):
    z=np.where(y_train1==label)
    temp=X_train1.iloc[np.where(y_train1==label)]
    y=(temp-xtest)**2
    ss=[]
    for i in range(len(y)):
        ss.append(np.sqrt(sum(y.iloc[i])))
        
    for i in range(len(y)):
        minimum[z[0][i]]=min(minimum[z[0][i]],ss[i])
    minimum.append(min(ss))
    return(np.array(minimum))
```

```python
def diff_label1(minimum,xtest,label):
    z=np.where(y_train1!=label)
    temp=X_train1.iloc[np.where(y_train1!=label)]
    y=(temp-xtest)**2
    ss=[]
    for i in range(len(y)):
        ss.append(np.sqrt(sum(y.iloc[i])))
        
    for i in range(len(y)):
        minimum[z[0][i]]=min(minimum[z[0][i]],ss[i])
    minimum.append(min(ss))
    return(np.array(minimum))
```

```python
np.seterr(divide='ignore', invalid='ignore')
```

**CALCULATING P_VALUE**

```python
p_value1=np.zeros((len(X_test1),len(label)))
for i in range(len(X_test1)):
    for j in range(len(label)):
        mini=mini1.copy()
        c4=same_label1(mini,X_test1.iloc[i],label[j])
        mini2=mini3.copy()
        c3=diff_label1(mini2,X_test1.iloc[i],label[j])
        c_score1=c3/c4
        p1=rankdata(c_score1,method='max')
        p_value1[i][j]=p1[-1]/(len(X_train1)+1)
```

**P_VALUE DATAFRAME FOR LABELS**

```python
df4=pd.DataFrame(p_value1,columns=label)
df4
```
